### Sigmoid function

Let's implement the sigmoid function for a vector x using numpy (vectorization) using the function [np.exp](https://docs.scipy.org/doc/numpy-1.10.1/reference/generated/numpy.exp.html) to implement it: 

$$ \text{For } x \in \mathbb{R}^n \text{,     } \sigma(x) = sigmoid(x) = sigmoid\begin{pmatrix}
    x_1  \\
    x_2  \\
    ...  \\
    x_n  \\
\end{pmatrix} = \begin{pmatrix}
    \frac{1}{1+e^{-x_1}}  \\
    \frac{1}{1+e^{-x_2}}  \\
    ...  \\
    \frac{1}{1+e^{-x_n}}  \\
\end{pmatrix} \tag{1} $$

Let's implement the tangent hyperbolic function for a vector x using numpy (vectorization) using the function np.tanh:

$$ \text{For } x \in \mathbb{R}^n \text{,     } tanh(x) = tanh\begin{pmatrix}
    x_1  \\
    x_2  \\
    ...  \\
    x_n  \\
\end{pmatrix} = \begin{pmatrix}
    tanh({x_1})  \\
    tanh({x_2})  \\
    ...  \\
    tanh({x_n})  \\
\end{pmatrix} \tag{2} $$

Let's implement the tangent hyperbolic function for a vector x using numpy (vectorization) using the function np.maximum:

$$ \text{For } x \in \mathbb{R}^n \text{,     } relu(x) = relu\begin{pmatrix}
    x_1  \\
    x_2  \\
    ...  \\
    x_n  \\
\end{pmatrix} = \begin{pmatrix}
    max(0, {x_1})  \\
    max(0, {x_2})  \\
    ...  \\
    max(0, {x_n})  \\
\end{pmatrix} \tag{3} $$

Let's implement the tangent hyperbolic function for a vector x using numpy (vectorization) using the function np.maximum:

$$ \text{For } x \in \mathbb{R}^n \text{,     } leakyrelu(x, \alpha) = leakyrelu\begin{pmatrix}
    x_1, \alpha  \\
    x_2, \alpha  \\
    ...  \\
    x_n, \alpha  \\
\end{pmatrix} = \begin{pmatrix}
    max({x_1} * \alpha, {x_1})  \\
    max({x_2} * \alpha, {x_2})  \\
    ...  \\
    max({x_n} * \alpha, {x_n})  \\
\end{pmatrix} \tag{4} $$

Let's implement the tangent hyperbolic function for a vector x using numpy (vectorization) using the function np.log and np.exp:

$$ \text{For } x \in \mathbb{R}^n \text{,     } softplus(x) = softplus\begin{pmatrix}
    x_1  \\
    x_2  \\
    ...  \\
    x_n  \\
\end{pmatrix} = \begin{pmatrix}
    log(1+e^{x_1})  \\
    log(1+e^{x_2})  \\
    ...  \\
    log(1+e^{x_n})  \\
\end{pmatrix} \tag{5} $$

```python
import numpy as np

def sigmoid(x):
    s = 1 / (1 + np.exp(-x))   # use np.exp to implement sigmoid activation function that works on a vector or a matrix
    return s

def tanh(x):
    t = np.tanh(x)   # np.tanh to implement tanh activation function that works on a vector or a matrix
    return t

def relu(x):
    r = np.maximum(0, x)   # use np.maximum to implement relu activation function that works on a vector or a matrix
    return r

def leakyrelu(x, slope):
    l = np.maximum(x * slope, x)   # use np.maximum to implement leaky relu activation function that works on a vector or a matrix
    return l

def softplus(x):
    p = np.log(1 + np.exp(x))   # use np.log and np.exp to implement softplus activation function that works on a vector or a matrix
    return p
```

```python
import matplotlib.pyplot as plt   # We will use the matplotlib to show the results

x = np.array([-4, -3.5, -3, -2.5, -2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4])
s = sigmoid(x)
t = tanh(x)
r = relu(x)
l = leakyrelu(x, 0.1)
p = softplus(x)
fig, ax = plt.subplots(figsize=(14,8))
sig_ = ax.plot(x, s)
tanh_ = ax.plot(x, t)
relu_ = ax.plot(x, r)
leakyrelu_ = ax.plot(x, l)
softplus_ = ax.plot(x, p)
ax.legend(['sigmoid', 'tanh', 'relu', 'leakyrelu', 'softplus'])
plt.show()
```

### Gradient functions

We need to compute gradients to optimize loss (error) function L using the backpropagation algorithm.

Let's implement the function sigmoid_derivative() to compute the gradient of the sigmoid function with respect to its input x using the following formula: 

$$sigmoid\_derivative(x) = \sigma'(x) = \sigma(x) (1 - \sigma(x)) = \begin{pmatrix}
    \sigma(x_1) (1 - \sigma(x_1))  \\
    \sigma(x_2) (1 - \sigma(x_2))  \\
    ...  \\
    \sigma(x_n) (1 - \sigma(x_n))  \\
\end{pmatrix} \tag{6}$$


Let's implement the function tanh_derivative() to compute the gradient of the tangent hyperbolic function with respect to its input x using the following formula: 

$$tanh\_derivative(x) = tanh'(x) = 1 - \sigma(x)^2 = \begin{pmatrix}
    1 - \sigma(x_1)^2  \\
    1 - \sigma(x_2)^2  \\
    ...  \\
    1 - \sigma(x_n)^2  \\
\end{pmatrix}  \tag{7}$$


Let's implement the function relu_derivative() to compute the gradient of the ReLu function with respect to its input x using the following formula: 

$$relu\_derivative(x) = relu'(x) = \begin{pmatrix}
    if \: x_1 > 0: 1 \: else \: 0 \\
    if \: x_2 > 0: 1 \: else \: 0 \\
    ...  \\
    if \: x_n > 0: 1 \: else \: 0 \\
\end{pmatrix}  \tag{8}$$


Let's implement the function leakyrelu_derivative() to compute the gradient of the leaky ReLu function with respect to its input x using the following formula: 

$$leakyrelu\_derivative(x, \alpha) = leakyrelu'(x, \alpha) = \begin{pmatrix}
    if \: x_1 > 0: 1 \: else \: \alpha \\
    if \: x_2 > 0: 1 \: else \: \alpha \\
    ...  \\
    if \: x_n > 0: 1 \: else \: \alpha \\
\end{pmatrix}  \tag{9}$$


Let's implement the function softplus_derivative() to compute the gradient of the soft plus function with respect to its input x using the following formula: 

$$softplus\_derivative(x) = softplus'(x) = \sigma(x) (1 - \sigma(x)) = \begin{pmatrix}
    \sigma(x_1) (1 - \sigma(x_1))  \\
    \sigma(x_2) (1 - \sigma(x_2))  \\
    ...  \\
    \sigma(x_n) (1 - \sigma(x_n))  \\
\end{pmatrix} = \begin{pmatrix}
    \frac{1}{1+e^{-x_1}}  \\
    \frac{1}{1+e^{-x_2}}  \\
    ...  \\
    \frac{1}{1+e^{-x_n}}  \\
\end{pmatrix} \tag{10}$$

```python
def sigmoid_derivative(x):
    s = sigmoid(x)
    dLs = s * (1 - s)
    return dLs

def tanh_derivative(x):
    t = tanh(x)
    dLt = 1 - t * t
    return dLt

def relu_derivative(x):
    r = relu(x)
    dLr = np.heaviside(x, 0)
    return dLr

def leakyrelu_derivative(x, slope):
    l = leakyrelu(x, slope)
    dLl = np.ones_like(x)
    dLl[x < 0] = slope
    return dLl

def softplus_derivative(x):
    p = softplus(x)
    dLp = 1 / (1 + np.exp(-x))
    return dLp
```

```python
x = np.array([-10, -8, -6, -4, -3, -2, -1, 0, 1, 2, 3, 4, 6, 8, 10], dtype=float)
print ("sigmoid_derivative(x) = " + str(sigmoid_derivative(x)))
print ("tanh_derivative(x) = " + str(tanh_derivative(x)))
print ("relu_derivative(x) = " + str(relu_derivative(x)))
print ("leakyrelu_derivative(x, slope = 0.05) = " + str(leakyrelu_derivative(x, 0.05)))
print ("softplus_derivative(x) = " + str(softplus_derivative(x)))
```

### Normalizing rows for better convergence efficiency

We usually normalize data to achieve better performance in machine learning because gradient descent converges faster after normalization. By normalization we mean changing x to $ \frac{x}{\| x\|} $ where $\| x\| = \sqrt{\sum_i{x_i^2}}$, i.e. dividing each row vector of x by its norm as follows:

If $$x = 
\begin{bmatrix}
    3 & 6 & 2 \\
    4 & 0 & 3 \\
\end{bmatrix}\tag{11}$$

then $$\| x\| = np.linalg.norm(x, axis = 1, keepdims = True) = \begin{bmatrix}
    7 \\
    5 \\
\end{bmatrix}\tag{12}$$

and        $$ x\_normalized = \frac{x}{\| x\|} = \begin{bmatrix}
    \frac{3}{7} & \frac{6}{7} & \frac{2}{7} \\
    \frac{4}{5} & \frac{0}{5} & \frac{3}{5} \\
\end{bmatrix}\tag{13}$$ 

We use here broadcasting that uses the same values of the computed norms to devide all values in rows.

```python
def normalizeRows(x):
    x_norm = np.linalg.norm(x, ord=2, axis=1, keepdims=True)  # norm of the 2nd order
    x = x/x_norm
    return x
```

```python
x = np.array([[3, 6, 2],
              [4, 0, 3]])
print("normalizeRows(x) = " + str(normalizeRows(x)))
```

### Softmax function ####
We often use a softmax function when classifying object to two or more classes. The softmax function normalizes the outputs by the row-sum of computed values:

**Instructions**:
- $ \text{For a vector} x \in \mathbb{R}^{1\times n} \text{,     } softmax(x) = softmax(\begin{bmatrix}
    x_1  &&
    x_2 &&
    ...  &&
    x_n  
\end{bmatrix}) = \begin{bmatrix}
     \frac{e^{x_1}}{\sum_{j}e^{x_j}}  &&
    \frac{e^{x_2}}{\sum_{j}e^{x_j}}  &&
    ...  &&
    \frac{e^{x_n}}{\sum_{j}e^{x_j}} 
\end{bmatrix} $ 

- $\text{For a matrix } x \in \mathbb{R}^{m \times n} \text{,  $x_{ij}$ maps to the element in the $i^{th}$ row and $j^{th}$ column of $x$, thus we have: }$  $$softmax(x) = softmax\begin{bmatrix}
    x_{11} & x_{12} & \dots  & x_{1n} \\
    x_{21} & x_{22} & \dots  & x_{2n} \\
    \vdots & \vdots & \ddots & \vdots \\
    x_{m1} & x_{m2} & \dots  & x_{mn}
\end{bmatrix} = \begin{bmatrix}
    \frac{e^{x_{11}}}{\sum_{j}e^{x_{1j}}} & \frac{e^{x_{12}}}{\sum_{j}e^{x_{1j}}} & \dots  & \frac{e^{x_{1n}}}{\sum_{j}e^{x_{1j}}} \\
    \frac{e^{x_{21}}}{\sum_{j}e^{x_{2j}}} & \frac{e^{x_{22}}}{\sum_{j}e^{x_{2j}}} & \dots  & \frac{e^{x_{2n}}}{\sum_{j}e^{x_{2j}}} \\
    \vdots & \vdots & \ddots & \vdots \\
    \frac{e^{x_{m1}}}{\sum_{j}e^{x_{mj}}} & \frac{e^{x_{m2}}}{\sum_{j}e^{x_{mj}}} & \dots  & \frac{e^{x_{mn}}}{\sum_{j}e^{x_{mj}}}
\end{bmatrix} = \begin{pmatrix}
    softmax\text{(1st row of x)}  \\
    softmax\text{(2nd row of x)} \\
    ...  \\
    softmax\text{(last row of x)} \\
\end{pmatrix} \tag{14}$$

```python
def softmax(x):
    x_exp = np.exp(x)
    x_sum = np.sum(x_exp, axis=1, keepdims=True)
    s = x_exp/x_sum   # x_sum is broadcasted here
    return s
```

```python
x = np.array([
    [0.9, 0.2, 0.8, 0.1, 1.0],
    [0.7, 0.5, 0.3, 0.0, 0.2]])
print("softmax(x) = " + str(softmax(x)))
```

### Implementation of the L1, L2, and L3 loss functions

The loss functions are used to evaluate the performance of the model. The bigger your loss is, the more different your predictions ($ \hat{y} $) are from the desired values ($y$). In machine learning and deep learning, you use optimization algorithms like gradient descent to train the model in such a way to minimize the cost function that consists of the losses for all training objects:

**L1 loss is defined as:**:
$$\begin{align*} & L_1(\hat{y}^{(i)}, y^{(i)}) = \sum_{j=0}^n|y_j^{(i)} - \hat{y}_j^{(i)}| \end{align*}\tag{15}$$

**L2 loss is defined as:**
$$\begin{align*} & L_2(\hat{y}^{(i)},y^{(i)}) = \sum_{j=0}^n(y_j^{(i)} - \hat{y}_j^{(i)})^2 \end{align*}\tag{16}$$

**L3 loss is defined as:**
$$\begin{align*} & L_3(\hat{y}^{(i)},y^{(i)}) = - \sum_{j=0}^n(y_j^{(i)} \log(\hat{y}_j^{(i)}) + (1 - y_j^{(i)}) (1 - \log(\hat{y}_j^{(i)}))) \end{align*}\tag{17}$$

```python
def L1(yhat, y):
    loss = np.sum(np.abs(y-yhat))
    return loss

def L2(yhat, y):
    loss = np.sum(np.dot(y-yhat,y-yhat))
    return loss

def L3(yhat, y):
    loss = np.sum(np.dot(y, np.log(yhat)) + np.dot((1 - y),(1 - np.log(yhat))))
    return loss
```

```python
yhat = np.array([0.9, 0.2, 0.1, .4, .9])
y = np.array([1, 0, 0, 0, 1])
print("L1 = " + str(L1(yhat,y)))

yhat = np.array([0.9, 0.2, 0.1, .4, .9])
y = np.array([1, 0, 0, 0, 1])
print("L2 = " + str(L2(yhat,y)))

yhat = np.array([0.9, 0.2, 0.1, .4, .9])
y = np.array([1, 0, 0, 0, 1])
print("L3 = " + str(L3(yhat,y)))
```
